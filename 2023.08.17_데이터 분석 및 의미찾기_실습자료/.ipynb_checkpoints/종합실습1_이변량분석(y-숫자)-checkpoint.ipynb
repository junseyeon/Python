## **2.숫자형 X --> Y**

* 모든 숫자형 X에 대해서 Y와 비교하여 차트를 그리고 수치화 하시오.

① 시각화 : scatter, jointplot  
② 수치화 : 상관분석

### **(1) Advertising -> Sales**

1) 시각화 : scatter(regplot), jointplot

```python
var = 'Advertising'

sns.scatterplot(x=var, y = target, data = data)
plt.show()
```

```python
sns.regplot(x=var, y = target, data = data)
plt.show()
```

```python
sns.jointplot(x=var, y = target, data = data)
plt.show()
```

2) 수치화 : 상관분석

```python
result = spst.pearsonr(data[var], data[target])
print(f'상관계수 : {result[0]}, p-value : {result[1]}')
```

3) 시각화와 수치화 결과로부터 파악한 내용을 적어 봅시다.

* 광고비와 판매량은 관련이 있다. 그러나 크지 않다.

* 추가 : 광고비 0 빼고 분석해보기

```python
temp = data.loc[data['Advertising'] > 0]
sns.scatterplot(x=var, y = target, data = temp)
plt.show()
```

```python
result = spst.pearsonr(temp[var], temp[target])
print(f'상관계수 : {result[0]}, p-value : {result[1]}')
```

### **(2) Urban -> Sales**

1) 시각화 : 평균비교 barplot

```python
var = 'Urban'
sns.barplot(x=var, y=target, data=data)
```

2) 수치화 : t-test, anova

```python
y= data.loc[data[var]=='Yes',target]
n = data.loc[data[var]=='No', target]

spst.ttest_ind(y,n)
```

3) 시각화와 수치화 결과로부터 파악한 내용을 적어 봅시다.

대립가설이 맞지 않다고 볼수 있음. ttest결과도 2보다 작고 pvalue도 1에 가까움으로 관계가 없다.

### **(3) US-> Sales**

1) 시각화 : 평균비교 barplot

```python
var = 'US'
sns.barplot(x=var, y=target, data=data)
```

2) 수치화 : t-test, anova

```python
usy = data.loc[data[var]=='Yes', target]
usn = data.loc[data[var]=='No', target]
spst.ttest_ind(usy, usn)
```

3) 시각화와 수치화 결과로부터 파악한 내용을 적어 봅시다.
